# Task-4: To Explore Decision Tree Algorithm

### Importing Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the dataset

```
dataset = pd.read_csv('Iris.csv')
dataset.head()
```

### Label Encoding

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Species'] = le.fit_transform(dataset['Species'])
dataset.head()
```

### Getting the features and label/target

```
x = dataset.iloc[:,1:5].values
y = dataset.iloc[:, 5].values
```

### Visualizing the dataset

```
#Swarmplot
sns.swarmplot(x = 'Species', y = 'SepalLengthCm', data = dataset)
```

```
#barplot
sns.barplot(x = 'SepalLengthCm', y = 'Species', data = dataset, estimator = np.mean)
```

```
#Catplot
sns.catplot(x = 'Species', y = 'PetalWidthCm', data = dataset)
```

```
#Catplot
sns.catplot(x = 'PetalLengthCm', y = 'Species', data = dataset, estimator=np.mean)
```

```
# jointplot
sns.jointplot(x = 'PetalLengthCm', y = 'Species', data = dataset, color = 'blue', kind = 'kde')
```

### Splitting the dataset

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Training the Decision tree algorithm

```
from sklearn.tree import DecisionTreeClassifier
classifier =  DecisionTreeClassifier()
classifier.fit(x_train, y_train)
```

### Accuracy and Confusion matrix for given model

```
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, classifier.predict(x_test))
cm = confusion_matrix(y_test, classifier.predict(x_test))
print('Accuracy of model is: ', accuracy)
sns.heatmap(cm, annot = True, fmt = 'd')
```

### Visualising the decision tree plot

```
from sklearn.tree import plot_tree
dtree = DecisionTreeClassifier().fit(x,y)
plt.figure(figsize = (25,20))
plot_tree(dtree, rounded = True, filled = True)
plt.show()
```
