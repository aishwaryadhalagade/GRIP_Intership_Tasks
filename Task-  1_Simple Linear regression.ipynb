# To Explore Supervised Machine Learning

### Importing the Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Importing the dataset from url

```
data_url = 'http://bit.ly/w-data'
dataset = pd.read_csv(data_url)
dataset
```

Linear relationship between Hours student studied and percentage score is shown by plot following:

### Visualising the dataset

```
dataset.plot(x = 'Hours', y = 'Scores', style = 'o')
plt.title('Hours vs. Percentage score')
plt.xlabel('Hours student studied')
plt.ylabel('Score in percentages')
plt.show()
```

### Splitting the dataset

```
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:, -1].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = 0)
```

### Training the data for Prediction

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predicting the Test set data

```
y_pred = regressor.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Visualising the training data

```
plt.scatter(X_train,y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Prediction with training data')
plt.show()
```

```
plt.scatter(X_test,y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('Prediction with test data')
plt.show()
```
